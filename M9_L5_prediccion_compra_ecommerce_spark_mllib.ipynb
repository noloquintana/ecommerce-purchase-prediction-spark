## 🎯 Objetivo  
El objetivo de este proyecto es desarrollar un modelo de Machine Learning supervisado que permita predecir la probabilidad de que un cliente realice una compra en un sitio de e-commerce, utilizando datos de comportamiento de navegación y consumo.

## 🧠 Metodología  
Se aplicarán las siguientes etapas del proceso de Machine Learning:

- Preprocesamiento y limpieza de datos  
- Transformación de variables y vectorización de características  
- División de datos en entrenamiento y prueba  
- Entrenamiento de modelos de clasificación (Regresión Logística o Random Forest)  
- Ajuste de hiperparámetros mediante validación cruzada  
- Evaluación del modelo utilizando métricas como precisión y AUC-ROC  

## 📁 Dataset  
Se utilizará el dataset **Online Shoppers Purchasing Intention**, el cual contiene información sobre el comportamiento de usuarios en un sitio de e-commerce.  

Este dataset incluye variables como número de visitas, duración de sesión, páginas vistas y tipo de visitante, junto con una variable objetivo (**Revenue**) que indica si el usuario realizó una compra (1) o no (0).

## 1. Introducción  

En el contexto actual, las organizaciones generan grandes volúmenes de datos provenientes de múltiples fuentes, como navegación web, historial de compras y comportamiento de usuarios. Este crecimiento en la cantidad de información hace que los enfoques tradicionales de análisis resulten insuficientes, especialmente cuando se busca obtener valor en tiempo real y a gran escala.

En particular, las empresas de e-commerce enfrentan el desafío de comprender el comportamiento de sus clientes para anticipar decisiones de compra y personalizar sus estrategias de marketing. Esto implica analizar grandes cantidades de datos y construir modelos predictivos capaces de identificar patrones relevantes.

El aprendizaje supervisado permite abordar este tipo de problemas mediante el uso de datos etiquetados, donde el objetivo es predecir una variable específica. En este caso, se busca predecir si un usuario realizará una compra, lo que corresponde a un problema de clasificación.

Para desarrollar esta solución, se utilizará el dataset **Online Shoppers Purchasing Intention**, que contiene información sobre el comportamiento de usuarios en un sitio web, incluyendo variables como número de visitas, duración de sesión y tipo de visitante. La variable objetivo (**Revenue**) indica si el usuario realizó una compra.

Dado el gran volumen de datos y la necesidad de escalabilidad, se empleará **Apache Spark MLlib**, una biblioteca de Machine Learning distribuido que permite entrenar modelos de forma eficiente en entornos de Big Data.

Para abordar el problema, se implementarán modelos de clasificación supervisada, tales como:

- Regresión Logística  
- Random Forest  

El objetivo principal es construir un modelo predictivo capaz de estimar la probabilidad de compra de los usuarios, evaluando su desempeño mediante métricas como precisión y AUC-ROC.

Finalmente, se analizarán los resultados obtenidos para determinar la efectividad del modelo y su potencial aplicación en la toma de decisiones dentro de la empresa.
